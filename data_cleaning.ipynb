```python
print("delete duplicate values")
me1df=medf.drop_duplicates()
print(me1df)
me1df.head(500)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(me1df)
```

```python
pca = PCA(n_components=None)
pca.fit(scaled_data)
```

```python
explained_variance=pca.explained_variance_ratio_
```

```python
components=pca.components_
```

```python
pca_df = pd.DataFrame(components,columns=me1df.columns)
explained_variance_df=pd.DataFrame(explained_variance,columns=['explained variance'])
```

```python
pca_df.shape
```

```python
scaled_data.shape
```

```python


print(pca_df)#PCA Components
```

```python
pca_df.head(500)
```

```python
print(explained_variance_df)#explained Variance Ratios
```

```python
abs_loadings = pca_df.abs()
print(abs_loadings)
```

```python
top_n = 2
top_features = {}
for i in range(abs_loadings.shape[1]):
 top_features[f'pc{i}'] = abs_loadings.iloc[:,i].nlargest(top_n).index.tolist()
 for pc, features in top_features.items():
  pass
 print(f"{pc}:{features}")
```

```python
cumulative_contributions = abs_loadings.sum(axis=1)
```

```python
s_c = cumulative_contributions.sort_values(ascending=False)
```

```python
print("cumulative_contributions")
print(s_c)
```

```python
most_imp_feature = s_c.idxmax()
h_c = s_c.max()
```

```python
print(f"\n most img feature: {most_imp_feature} with contribution: {h_c}")
```
